### Read the data from excel sheet

### Filling Missing Values of Age with mean

```python
data_frame.loc[data_frame["Age"].isna(), "Age"] = data_frame["Age"].mean()
```

```python
data_frame["Age"] = data_frame["Age"].astype(int)
```

```python
data_frame
```

### Removing unwanted characters from Salary

```python
data_frame["Salary"] = data_frame["Salary"].str.strip("%$")
```

### Handling missing values in Salary

```python
data_frame [data_frame["Salary"].isna()]
```

```python
data_frame["Salary"]=data_frame["Salary"].fillna(0).astype(int)
```

```python
data_frame
```

### Standardizing Categorical Values in Gender

```python
data_frame
```

Capitalizing the Values in Gender

```python
data_frame["Gender"] = data_frame["Gender"].str.title()
data_frame
```

### Handling Missing Values in Gender

```python
data_frame["Gender"] = data_frame["Gender"].fillna("Undefined")
```

```python
data_frame
```

### Splitting Type_of_Ownership Column

```python
data_frame
```

```python
data_frame["Type_of_Ownership"]=data_frame["Type_of_Ownership"].fillna("Undefined")
```

```python
data_frame["Privacy"] = data_frame["Type_of_Ownership"].str.split(pat = "-").str[1]
```

```python
data_frame["Type_of_Ownership"] = data_frame["Type_of_Ownership"].str.split(pat = "-").str[0]
```

```python
data_frame["Privacy"]=data_frame["Privacy"].fillna("Undefined")
```

 sorting "Type of Ownership" and "Privacy" together

```python
type_of_ownership = data_frame.pop('Type_of_Ownership')

data_frame.insert(data_frame.columns.get_loc('Privacy'), 'Type_of_Ownership', type_of_ownership)
```

```python
data_frame["Zip"] = data_frame["Zip"].replace("F" , "Undefined").fillna("Undefined")
data_frame["Contract_Type"] = data_frame["Contract_Type"].fillna("Undefined")
data_frame
```
